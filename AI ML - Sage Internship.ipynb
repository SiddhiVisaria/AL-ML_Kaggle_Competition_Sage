```python
emissions_quantity_sum = df1.groupby('iso3_country')['emissions_quantity'].sum().reset_index()
emissions_quantity_sum.columns = ['iso3_country', 'emissions_quantity_sum']
```

```python
df1['start_time'] = pd.to_datetime(df1['start_time'], format='%m/%d/%y %H:%M')
df1 = df1[(df1['start_time'].dt.year >= 2015) & (df1['start_time'].dt.year <= 2021)]
```

```python
df_pivot = df1.pivot_table(values='emissions_quantity', index='iso3_country', columns=df1['start_time'].dt.year)
co2_change = df_pivot[2021] - df_pivot[2015]
co2_change = co2_change.reset_index()
co2_change.columns = ['iso3_country', 'emissions_change']
```

```python
df_merged = pd.merge(emissions_quantity_sum, co2_change, on='iso3_country')
```

```python
scaler = StandardScaler()
df_merged['emissions_quantity_sum_norm'] = scaler.fit_transform(df_merged[['emissions_quantity_sum']])
df_merged['emissions_change_norm'] = scaler.fit_transform(df_merged[['emissions_change']])
```

```python
usa_row = df_merged[df_merged['iso3_country'] == 'USA'][['emissions_quantity_sum_norm', 'emissions_change_norm']]
df_merged['distance'] = euclidean_distances(df_merged[['emissions_quantity_sum_norm', 'emissions_change_norm']], usa_row)
```

```python
df_final = df_merged[['iso3_country', 'emissions_quantity_sum_norm', 'emissions_change_norm', 'distance']]
df_final.to_csv('output.csv', index=False)
```
